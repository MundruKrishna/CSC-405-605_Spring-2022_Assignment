```python
#importing CSV libraries which helps to read given TSV format list
#importing random libraries which helps to randomly assign the students to the teams list
import csv
import random



# Opening the file with the help of dynamic path to the folder
file = open("../Assignment1/sp-22-students.tsv")
givenData = csv.reader(file, delimiter='\t')
temp_list=next(givenData)


# collecting the data into a list called allStudents 
allStudents=[]
for x in givenData:
    # Removing the empty strings inside the list so that all the empty cells will get eliminated while printing the output
    x=[i for i in x if i]
    allStudents.append(x)
    #shuffling the students data randomly using random library imported
random.shuffle(allStudents)

#print(Jaya)

# Collecting the number of students in to the studentsNumber variable which can be used to assign further

studentsNumber=len(allStudents)

# Collecting the the variable for requirement of number of teams 

group=int(input('Please enter the requirement of number of teams : '))
for i in range (0,int(group)):
    allTeams=[]
    for t in range(i, studentsNumber, int(group)):
        allTeams.append(allStudents[t])
    print('\nTeam Members of the team ',i+1,' are:')
    print (*Teams,sep='\n')

```
